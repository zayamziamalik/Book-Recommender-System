```python
# import numpy as np
import pandas as pd
users=pd.read_csv("Users.csv")
books=pd.read_csv("Books.csv")
ratings=pd.read_csv("Ratings.csv")
```

# Popularity Based

# Q1)How to merge two tables in Pandas?

```python
ratings_with_name=ratings.merge(books, on="ISBN")
ratings_with_name
```

# Q2)How to group something on specific column in Pandas?

# Q3)How to reset index in Pandas?

# Q4)How to rename a column in Pandas? 

```python
num_ratings_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_ratings_df.rename(columns={'Book-Rating':'num_ratings'}, inplace=True)
mt_num_ratings=num_ratings_df
num_ratings_df
```

# Q5)How to convert a column data to numeric with Pandas?

# Q6)How to take average of values of a column on basis of a specific column for its values?

```python
import numpy as np
ratings_with_name['Book-Rating'] = pd.to_numeric(ratings_with_name['Book-Rating'], errors='coerce')
avg_ratings_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_ratings_df.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)
avg_ratings_df
```

```python
popular_df=num_ratings_df.merge(avg_ratings_df,on='Book-Title')
popular_df
```

# Q7)How to sort dataset on basis on one column in Pandas?

# Q8)How to remove Duplicate data on basis of one column in Pandas?

# Q9)How to extract first n number of rows after performing some operations?

```python
popular_df=popular_df[popular_df["num_ratings"]>=250].sort_values('avg_rating', ascending=False).head(50)
popular_df=popular_df.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'Image-URL-M', 'num_ratings', 'avg_rating']]
```

# Collaborative

# Q10)How to count users who rated more than 200 books?

```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating']  >200
padhe_likhe_users=x[x].index
mt_ana=ratings_with_name.groupby('User-ID').count()['Book-Rating']
```

```python
filtered_rating=ratings_with_name[ratings_with_name["User-ID"].isin(padhe_likhe_users)]
```

# Q11)How to count books with more than 50 ratings?

```python
y=filtered_rating.groupby("Book-Title").count()['Book-Rating']>=50
famous_books=y[y].index
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
for_content=final_ratings
```

# Q12)How to pivot a table with Pandas?

# Q13)How to replace 'nan' values with '0' in Pandas?

```python
pt=final_ratings.pivot_table(index='Book-Title', columns="User-ID", values="Book-Rating")
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores=cosine_similarity(pt)
```

```python
def recommend(book_name):
    index=np.where(pt.index==book_name)[0][0]
    distances=similarity_scores[index]
    similar_items=sorted(list(enumerate(similarity_scores[index])), key=lambda x:x[1], reverse=True)[0:6]
    data=[]
    for i in similar_items:
        item=[]
        print(pt.index[i[0]])
        temp_df=books[books["Book-Title"]==pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Title"].values))
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Author"].values))
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Image-URL-M"].values))
        
        data.append(item)
    return data
```

```python
recommend("Message in a Bottle")
```

```python
import pickle
pickle.dump(pt, open('pt.pkl', 'wb'))
pickle.dump(books, open('books.pkl', 'wb'))
pickle.dump(similarity_scores, open('similarity_score.pkl', 'wb'))
pickle.dump(popular_df, open('popular.pkl', 'wb'))
```

# Content Based

# Q14)How to merge multiple columns into one column?

```python
books2=for_content[["ISBN", "Book-Title", "Book-Author", "Publisher"]].drop_duplicates("Book-Title")
books2["tags"]=books2["Book-Title"]+" "+books2["Book-Author"]+" "+books2["Publisher"]
books2=books2[["ISBN", "Book-Title", "tags"]]
books2['tags'] = books2['tags'].fillna('').astype(str).str.lower()
books2=books2.reset_index()
books2
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return" ".join(y)

books2["tags"]=books2["tags"].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
import numpy as np
vector = cv.fit_transform(books2['tags']).toarray()
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

# Q15)How to locate value by index using Pandas?

```python
def recommend2(movie):
    index = books2[books2['Book-Title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        item=[]
        temp_df = books[books['Book-Title'] == books2.iloc[i[0]]["Book-Title"]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
#         data.append(item)
        print(item)
        print(books2.iloc[i[0]]["Book-Title"])
        
```

```python
recommend2("1984")
```
